```python
import pandas as pd

# Load the CSV file into a DataFrame
file_path = 'diabetes_012_health_indicators_BRFSS2015.csv'
diabetes_data = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(diabetes_data.head())
```

```python
# Remove specified columns
columns_to_remove = [
    "Smoker", "PhysActivity", "Fruits", "Veggies",
    "HvyAlcoholConsump", "AnyHealthcare", "NoDocbcCost",
    "MentHlth", "PhysHlth", "DiffWalk", "Education", "Income"
]

# Drop the columns and create a new DataFrame
cleaned_data = diabetes_data.drop(columns=columns_to_remove)

# Display the first few rows of the cleaned DataFrame
print(cleaned_data.head())
```

```python
# Count the total number of rows using shape
total_rows = cleaned_data.shape[0]

# Display the result
print("Total number of participants=", total_rows)
```

```python
# Count the number of rows where "Sex" equals 0.0
count_sex_0 = (cleaned_data["Sex"] == 0.0).sum()

# Display the result
print("Number of participants who are Female=", count_sex_0)
```

```python
# Count the number of participants with Diabetes (Diabetes_012 == 1.0) and Sex == 0.0
count_diabetes_sex_0 = ((diabetes_data["Diabetes_012"] == 1.0) & (diabetes_data["Sex"] == 0.0)).sum()

# Display the result
print("Number of participants with Diabetes who are Female=", count_diabetes_sex_0)
```

```python
# Count the number of participants with Diabetes (Diabetes_012 == 1.0) and Sex == 1.0
count_diabetes_sex_1 = ((diabetes_data["Diabetes_012"] == 1.0) & (diabetes_data["Sex"] == 1.0)).sum()

# Display the result
print("Number of participants with Diabetes who are Male=", count_diabetes_sex_1)
```

```python
# Count the number of rows where "Diabetes_012" equals 1.0 and "HighBP" equals 1.0
count_diabetes_highbp = ((diabetes_data["Diabetes_012"] == 1.0) & (diabetes_data["HighBP"] == 1.0)).sum()

# Display the result
print("Number of participants with Diabetes who have High Blood Pressure=", count_diabetes_highbp)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Load the CSV file into a DataFrame
file_path = 'diabetes_012_health_indicators_BRFSS2015.csv'
diabetes_data = pd.read_csv(file_path)
 
# First pie chart - Overall diabetes distribution
diabetes_count = ((diabetes_data["Diabetes_012"] == 1.0) | (diabetes_data["Diabetes_012"] == 2.0)).sum()
total_rows = len(diabetes_data)
 
plt.figure(figsize=(8, 5))
plt.pie([diabetes_count, total_rows - diabetes_count], 
        explode=(0.1, 0),
        labels=['Participants with Pre/Diabetes', 'No Diabetes'],
        colors=['orange', 'lightblue'],
        autopct='%1.1f%%', 
        shadow=True, 
        startangle=140)
plt.title('Overall Percentage of Participants with Pre/Diabetes')
plt.axis('equal')
plt.show()
 
# Second pie chart - Female diabetes distribution
female_data = diabetes_data[diabetes_data["Sex"] == 0.0]
female_diabetes = ((female_data["Diabetes_012"] == 1.0) | (female_data["Diabetes_012"] == 2.0)).sum()
total_females = len(female_data)
 
plt.figure(figsize=(8, 5))
plt.pie([female_diabetes, total_females - female_diabetes],
        labels=['Females with Pre/Diabetes', 'Females without Diabetes'],
        autopct='%1.1f%%',
        startangle=90,
        colors=['orange', 'lightblue'])
plt.title('Pre/Diabetes Distribution Among Female Participants')
plt.axis('equal')
plt.show()
 
# Third pie chart - Male diabetes distribution
male_data = diabetes_data[diabetes_data["Sex"] == 1.0]
male_diabetes = ((male_data["Diabetes_012"] == 1.0) | (male_data["Diabetes_012"] == 2.0)).sum()
total_males = len(male_data)
 
plt.figure(figsize=(8, 5))
plt.pie([male_diabetes, total_males - male_diabetes],
        labels=['Males with Pre/Diabetes', 'Males without Diabetes'],
        autopct='%1.1f%%',
        startangle=90,
        colors=['orange', 'lightblue'])
plt.title('Pre/Diabetes Distribution Among Male Participants')
plt.axis('equal')
plt.show()
 
# Print the actual numbers for verification
print(f"\nTotal participants: {total_rows}")
print(f"Total with pre/diabetes: {diabetes_count} ({diabetes_count/total_rows*100:.1f}%)")
print(f"\nFemale participants: {total_females}")
print(f"Females with pre/diabetes: {female_diabetes} ({female_diabetes/total_females*100:.1f}%)")
print(f"\nMale participants: {total_males}")
print(f"Males with pre/diabetes: {male_diabetes} ({male_diabetes/total_males*100:.1f}%)")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
# Load the data
df = pd.read_csv('diabetes_012_health_indicators_BRFSS2015.csv')
 
# Create figure for all visualizations
plt.figure(figsize=(15, 10))
 
# 1. High Blood Pressure Distribution
plt.subplot(2, 2, 1)
bp_counts = df['HighBP'].value_counts()
plt.pie([bp_counts[1], bp_counts[0]], 
        labels=['High BP', 'Normal BP'],
        colors=['red', 'lightblue'],
        autopct='%1.1f%%',
        startangle=90)
plt.title('Distribution of High Blood Pressure in Population')
 
# 2. High Cholesterol Distribution
plt.subplot(2, 2, 2)
chol_counts = df['HighChol'].value_counts()
plt.pie([chol_counts[1], chol_counts[0]], 
        labels=['High Cholesterol', 'Normal Cholesterol'],
        colors=['orange', 'lightgreen'],
        autopct='%1.1f%%',
        startangle=90)
plt.title('Distribution of High Cholesterol in Population')
 
# 3. Age distribution for people with high BP
plt.subplot(2, 2, 3)
age_bp = df[df['HighBP'] == 1]['Age'].value_counts().sort_index()
plt.bar(range(len(age_bp)), age_bp, color='red', alpha=0.6)
plt.title('Age Distribution of People with High BP')
plt.xlabel('Age Group')
plt.ylabel('Count')
 
# 4. Age distribution for people with high cholesterol
plt.subplot(2, 2, 4)
age_chol = df[df['HighChol'] == 1]['Age'].value_counts().sort_index()
plt.bar(range(len(age_chol)), age_chol, color='orange', alpha=0.6)
plt.title('Age Distribution of People with High Cholesterol')
plt.xlabel('Age Group')
plt.ylabel('Count')
 
plt.tight_layout()
plt.show()
 
# Print some statistics
print("\nHigh Blood Pressure Statistics:")
print(f"Total people with high BP: {bp_counts[1]:,}")
print(f"Percentage of population: {(bp_counts[1]/len(df)*100):.1f}%")
 
print("\nHigh Cholesterol Statistics:")
print(f"Total people with high cholesterol: {chol_counts[1]:,}")
print(f"Percentage of population: {(chol_counts[1]/len(df)*100):.1f}%")
 
# Create a plot showing the relationship between BMI and these conditions
plt.figure(figsize=(12, 6))
 
# BMI distribution for different conditions
plt.subplot(1, 2, 1)
sns.boxplot(x='HighBP', y='BMI', data=df)
plt.title('BMI Distribution by Blood Pressure Status')
plt.xlabel('Has High Blood Pressure')
plt.ylabel('BMI')
 
plt.subplot(1, 2, 2)
sns.boxplot(x='HighChol', y='BMI', data=df)
plt.title('BMI Distribution by Cholesterol Status')
plt.xlabel('Has High Cholesterol')
plt.ylabel('BMI')
 
plt.tight_layout()
plt.show()
```
